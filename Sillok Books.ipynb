```python
import xml.etree.ElementTree as et
import glob
import pandas as pd
```

```python
booklist = []

for filename in glob.glob('sillok/*.xml'):
    tree = et.parse(filename)
    root = tree.getroot()
    entries = tree.findall('//level5')
    for entry in entries:
        entrydate = entry.find('front/biblioData/date/dateOccured')
        
        if entrydate is None:
            entrydate = "_"
        else:
            entrydate = entrydate.attrib['date']
        
        books = entry.findall('text/content/paragraph/index[@type="서명"]')
        if books is None:
            book = "_"
        else:
            for book in books:
                book = book.text
                entry = [filename, entrydate, book]
                print(entry)
                booklist.append(entry)

df = pd.DataFrame(booklist)
df.to_csv('sillok/sillok_books.csv', encoding='utf-8', sep='\t')
```
